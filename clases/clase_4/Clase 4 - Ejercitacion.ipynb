# Patrones de movilidad de ecobici

Como la enorme cantidad de informacion desagregada de viajes individuales entre estaciones puede ser muy compleja de comprender, podemos recurrir al recurso de la agregación.

Para esto se agrega un dataset de estacione donde cada estacino tiene el codigo del radio censal. Por lo tanto se intenta dar un soporte cartografico para agregar información. A continuación uds pueden:
